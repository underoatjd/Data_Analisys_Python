```python
# importamos pandas y numpy para poder leer y ejecutar nuestro archivo
import pandas as pd
import numpy as np

# definimos el nombre de nuestro dataset
dataset = pd.read_csv('U4_01_housing.csv')
#verificamos que nuestro dataset pueda ser leido
print(dataset.head())
print(dataset.info())
```

```python
#observamos la columna "ocean_proximity" porque en el resultado anterior nos arrojo que es un objeto y no un valor. 
dataset["ocean_proximity"].value_counts()
```

```python
#esta linea de codigo nos permite verificar informacion que describe los datos.
dataset.describe()
```

```python
#usamos esta linea de codigo para crear un histograma de las variables que pueden ser utilizadas en el histograma
dataset.hist(figsize=(15,8),bins=50,edgecolor='k')
```

```python
#importamos seaborn() paa poder graficar de forma llamativa el set de datos. 
import seaborn as sb
datamap = sb.scatterplot(x="latitude",y="longitude",data=dataset, hue="median_house_value", palette="coolwarm", size=dataset["population"]/100)
```

```python
# este es el mismo codigo de arriba pero podemos utilizar data=dataset[(dataset.*******)] para dar un filtro o condicion
datamap2 = sb.scatterplot(x="latitude",y="longitude",data=dataset[(dataset.median_income > 14)], hue="median_house_value", palette="coolwarm", size=dataset["population"]/10)
```

```python
#Una vez mas verificamos la informacion que describe el dataset pero con respecto a los datos nulos
dataset.info()
```

```python
#utilizamos la funcion .dropna() para crear un nuevo dataset que ya este limpio de datos vacios.
datasetLimpio = dataset.dropna()
datasetLimpio.info()
```

```python
# ahora debemos convertir la caracteristica categorica a numerica o sea [("ocean_proximity")]
# utilizamos value_counts() para explorar las etiquetas de datos dentro de ["ocean_proximity"]
datasetLimpio["ocean_proximity"].value_counts()
```

```python
# utilizamos la función de pandas get_dummies() para pasar los datos de "ocean_proximity" a numericos
#tendremos que utilizar dtype=int para que nos retorne un dato de tipo entero
dummies = pd.get_dummies(datasetLimpio["ocean_proximity"], dtype=int)
```

```python
#podremos ejecutar una linea de codigo para validar los (dummies)
print(dummies)
```

```python
# al datasetLimpio le agregamos los dummies de "ocean_proximity" con un join
datasetFinal = datasetLimpio.join(dummies)
```

```python
#eliminamos la columna "ocean_proximity" con la funcion drop, pero hay que explicarle que
# es una columna = [axis=1] y no una fila [axis=2]
datasetFinal = datasetFinal.drop(["ocean_proximity"], axis=1)
```

```python
#volvemos a ejecutar un codigo para describir la informacion y validar que
# los counts en todas las celdas es igual a 20433
datasetFinal.describe()
```

```python
#ahora podemos continuar con el analisis y verificamos las correlaciones con la funcion corr()
datasetFinal.corr()
```

```python
# En el code cell anterior nos dimos cuenta que la informacion no es muy clara, por lo tanto
# no detectamos alguna relacion a imple vista, por tal motivo graficaremos estos datos para 
# verificar si podemos relacionarlos.
sb.set(rc={"figure.figsize":(18,10)})
sb.heatmap(datasetFinal.corr(), annot=True, cmap="YlGnBu")
```

```python
#en esta celula de codigo podremos ver en orden las correlaciones de positivas a negativas.
datasetFinal.corr()["median_house_value"].sort_values(ascending=False)
```
